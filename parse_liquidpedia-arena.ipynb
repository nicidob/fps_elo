```python
import numpy as np
from itertools import combinations
from pysmt.shortcuts import Symbol, And, Not, is_sat, get_model, Or
from collections import defaultdict


def guess_group(s):
    func_res = []
    N = len(s)
    games = list(combinations(np.arange(0,N),2))
    g_idx = [i for i in range(len(games))]
    varL = [Symbol(str(i)) for i in g_idx]
    players = list(s.keys())
    records = [s[_] for _ in players]
    hopes = []
    for i,r in enumerate(records):
        p_games = [j for j,g in zip(g_idx,games) if i in g]
        if r > N//2:
            fr = r
            mul = False
        else:
            fr = len(p_games)-r
            mul = True
        combos = list(combinations(p_games,fr))
        #print(r, fr, combos)
        options = []
        for p in combos:
            lst = [varL[_] if games[_][0] ==i else Not(varL[_])  for _ in p]
            if mul:
                lst = [Not(_) for _ in lst]
            #print(And(lst))
            options.append(And(lst))
        hopes.append(Or(options))
    if is_sat(And(hopes)):
        mdl = get_model(And(hopes))
        res = [str(mdl.get_value(_)) == 'True' for _ in varL]
        for g,r in zip(games,res):
            if r:
                s1,s2 = g
            else:
                s2,s1 = g
            match = ('',players[s1],players[s2],2.5,0,[])
            func_res.append(match)
    return func_res
```

```python
results = {}
TB = {}
extras = {}
for k in data:
    if 'User:' in k:
        continue
    #print('\t\t',k)
    if k in bad_events:
        continue
    matches = []
    found_brack = False
    found_table = False
    team_lookup = {}
    clean_data = re.sub("(<!--.*?-->)", "", data[k], flags=re.DOTALL).replace('R1D1score2=','R1D1score=2').replace('R2W10score0','R2W10score=0').replace('R1D401score','R1D40score')
    clean_data = clean_data.replace('&#45;','-').replace('R2D6flaguk','R2D6flag')
    #clean_data = clean_data.replace('|R3W14=Slinger |R3W14score= |R3W14flag=de |R3W14win=','|R3W14=Slinger |R3W14score= |R3W14flag=de |R3W14win=1')
    wikicode = mwparserfromhell.parse(clean_data)
    place_count = 1
    place_list = []
    tier = 'U'
    onoff = 'Offline'
    potential_extras = []
    bad_group = False
    for template in wikicode.filter_templates():
        tn = str(template.name)
        tn = re.sub("(<!--.*?-->)", "", tn, flags=re.DOTALL)
        tn = tn.strip().rstrip()
        #rint(tn)
        if 'Infobox league'.lower() in tn.lower():
            if 'liquipediatier' in template:
                tier = template.get('liquipediatier').value.rstrip()
                tier = tier.replace('TierText','').replace('}','').replace('{','').replace('/','')
                #print(tier,template.get('liquipediatier'))
            if 'type' in template:
                onoff = template.get('type').value.rstrip()
            if 'sdate=' in template:
                try:
                    sdate = template.get('sdate').value.strip().rstrip()
                    sdate = dateutil.parser.parse(sdate)

                    edate = template.get('edate').value.strip().rstrip()
                    edate = dateutil.parser.parse(edate)

                    mdate = sdate + (edate-sdate)/2
                except:
                    mdate = edate
                
            else:
                mdate = template.get('date').value.rstrip()
                mdate = dateutil.parser.parse(mdate.replace('??','01'))
            game = template.get('game').value.rstrip()
            #break
        elif 'MatchMaps' in tn:
            found_table = True
            if k not in bad_groups and len(str(template.get('player1').value).strip().rstrip()) > 0:
                if len(str(template.get('winner').value).strip().rstrip()) > 0 or len(str(template.get('games1').value).strip().rstrip()) > 0 or len(str(template.get('games2').value).strip().rstrip()) > 0 :
                    res = get_match(template)
                    if res:
                        matches.append(res)
                else:
                    print('uh oh ',k)
        elif tn in valid_matches:
            found_brack =True
            res = handle_brack(template,valid_matches[tn])
            matches = matches + res
        elif tn =='32DEFinalBracket':
            found_brack =True
            res = handle_brack(template,valid_matches['32DEBracket'])
            matches = matches + res
        elif tn == 'prize pool slot':
            names = [_ for _ in template.params if '=' not in _]
            place_count += len(names)
            if 'place' in template:
                place = template.get('place').value.split('-')[0]
            else:
                place = str(place_count)
            place = ''.join(place.split())
            #print(place,type(place),place.isdigit())
            if place.isdigit():
                place = int(place)
                #print(place)
                place_list.append((place,names))
        else:
            if 'GroupTableStart' == tn:
                group_list = {}
                bad_group = False
            elif 'GroupTableSlot' == tn:
                try:
                    wins = str(template.get('win_m').value)
                    player = template.params[0].replace('{','').replace('}','').split('|')[-1]
                    group_list[player] = int(wins)
                except:
                    print('bad group in ',k)
                    bad_group = True
            elif 'GroupTableEnd' == tn:
                if not bad_group and len(group_list) > 1:
                    if sum([int('team=' in k) for k in group_list]) == 0:
                        #print(group_list)
                        potential_extras  = potential_extras + guess_group(group_list)
                group_list = {}
            #print(tn)
            if 'Bracket' in tn and 'Summary' not in tn:
                if '32DEFinalBracket' == tn:
                    print('32B in ',k)
                filt = [_.split('=')[0].replace('flag','').replace('score','').replace('win','') for _ in template.params if 'score' in _]
                if (len(filt)%3) == 0:
                    rep =  sorted(get_tripl(filt),key=lambda x: x[2])
                    if tn in extras and len(rep) > len(extras[tn]):
                        extras[tn] = rep
                    else:
                        extras[tn] = rep
                elif (len(filt)%2) == 0:
                    rep =  sorted(get_dup(filt))
                    if tn in extras and len(rep) > len(extras[tn]):
                        extras[tn] = rep
                    else:
                        extras[tn] = rep
                
                TB[tn] = TB.get(tn,0) + 1
    if not found_brack:
        places_list = place_list[::-1]
        for idx, d in enumerate(places_list):
            pl, names = d
            to_lose_to = sum([_[1] for _ in places_list[idx+1:]],[])
            for l,w in zip(names,to_lose_to):
                matches.append(('',str(w),str(l),1.5,0,[]))
    if not found_table:
        matches = potential_extras  + matches
    #print(found_brack)
    results[k] = (tier,onoff,mdate,matches,game)
```
