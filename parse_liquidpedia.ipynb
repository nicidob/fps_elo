```python
def get_details(gm):
    try:
        gmt = gm.filter_templates()[0]
    except:
        return ('',[])
    gtime = ''
    try:
        gtime = dateutil.parser.parse(str(gmt.get('date').value).replace('Abbr/','').replace(',2',', ').replace('abbr/','').replace('{','').replace('}',''))
    except:
        pass
    rounds = []
    for k in range(7):
        try:
            mapn = gmt.get('map{}'.format(k+1)).value
            mapn = ''.join(mapn.split())
            tmpS = []
            if 'map{}score'.format(k+1) in gmt:
                scoreS = str(gmt.get('map{}score'.format(k+1)).value)
                tmpS = [re.sub('[^0-9]','', _) for _ in scoreS.split('-')]
                tmpS = [int(_) for _ in tmpS if len(_) > 0]

            if len(tmpS) != 2:   
                mapns1 = int(str(gmt.get('map{}t1t'.format(k+1)).value)) + int(str(gmt.get('map{}t1ct'.format(k+1)).value))
                mapns2 = int(str(gmt.get('map{}t2t'.format(k+1)).value)) + int(str(gmt.get('map{}t2ct'.format(k+1)).value))
                for otn in range(3):
                    ot1_str1 = 'map{}o{}t1ct'.format(k+1,otn)
                    ot1_str2 = 'map{}o{}t1t'.format(k+1,otn)
                    ot2_str1 = 'map{}o{}t2ct'.format(k+1,otn)
                    ot2_str2 = 'map{}o{}t2t'.format(k+1,otn)
                    if ot1_str1 in gmt and len(str(gmt.get(ot1_str1).value).rstrip()) > 0:
                        mapns1 += int(str(gmt.get(ot1_str1).value)) +  int(str(gmt.get(ot1_str2).value))
                        mapns2 += int(str(gmt.get(ot2_str1).value)) +  int(str(gmt.get(ot2_str2).value))
            else:
                mapns1,mapns2 = tmpS
            rounds.append((mapn,mapns1,mapns2))
        except:
            pass
    return (gtime, rounds)
```

```python
def get_match(templ):
    t1  = str(templ.get('team1').value)
    t2  = str(templ.get('team2').value)
    
    t1 = ''.join(t1.split())
    t2 = ''.join(t2.split())


    s1  = str(templ.get('games1').value).strip().rstrip().replace('FF','0').replace('W','1').replace('DQ','0').replace('<s>','').replace('</s>','')
    s2  = str(templ.get('games2').value).strip().rstrip().replace('FF','0').replace('W','1').replace('DQ','0').replace('<s>','').replace('</s>','')
    s1 =  re.sub('[^0-9]','', s1)
    s2 =  re.sub('[^0-9]','', s2)
    if len(s1) == 0 or len(s2) == 0:
        s1 = str(templ.get('winner').value).strip().rstrip()
        if len(s1) == 0:
            return None
        if 'draw' in s1 or 'draw' in s2:
            return None
        s1 = int(s1)
        s2 = 1 - s1
        s1 = str(s1)
        s2 = str(s2)
    if '?' in s1:
        s1 = 0
        s2 = 0
    s1 = int(s1)
    s2 = int(s2)

    if 'details' in templ:
        gm = templ.get('details').value
        gtime, rounds = get_details(gm)
    else:
        gtime, rounds = '', []
        
    if s1 + s2 >= 5:
        if len(rounds) == 1:
            r1 = rounds[0]
            rounds = [(r1[0],s1,s2)]
            s1 = int(rounds[0][1] > rounds[0][2])
            s2 = int(rounds[0][1] < rounds[0][2])

    if len(rounds) == 0 and s1 + s2 >= 5:
        mapt = ''
        try:
            gmt = gm.filter_templates()[0]
            mapt = str(gmt.get('map1','').value)
        except:
            pass
        rounds = [(mapt,s1,s2)]
        s1 = int(s1 > s2)
        s2 = 1-s1
    #print(t1,t2,s1,s2,rounds)
    return (gtime,t1,t2,s1,s2,rounds)
```

```python
import math
def handle_brack(templ,matchups):
    games = []
    for strb1,strb2, strg in matchups:
        t1s = strb1 + 'team'
        if t1s not in templ:
            t1s = strb1

        t2s = strb2 + 'team'
        if t2s not in templ:
            t2s = strb2
        if t1s in templ and t2s in templ:
            t1  = str(templ.get(t1s).value)
            t2  = str(templ.get(t2s).value)
            t1 = ''.join(t1.split())
            t2 = ''.join(t2.split())
            
            s1s = strb1 + 'score'
            s2s = strb2 + 'score'

            if s1s in templ and s2s in templ:
                s1  = str(templ.get(s1s).value).strip().rstrip().replace('FF','0').replace('W','1').replace('DQ','0')
                s2  = str(templ.get(s2s).value).strip().rstrip().replace('FF','0').replace('W','1').replace('DQ','0')

                s1 =  re.sub('[^0-9]','', s1)
                s2 =  re.sub('[^0-9]','', s2)
            elif strb1 + 'win' or strb2 + 'win' in templ:
                try:
                    s2 = int(str(templ.get(strb2 + 'win').value))
                    s1 = 1 - s2
                    s1, s2 = str(s1), str(s2)
                except:
                    s1 = int(str(templ.get(strb1 + 'win').value))
                    s2 = 1 - s1
                    s1, s2 = str(s1), str(s2)
            else:
                continue
            if 'draw' in s1 or 'draw' in s2:
                continue

            if len(s1) == 0:
                s1 = '0'
            if len(s2) == 0:
                s2 = '0'


            s1 = int(s1)
            s2 = int(s2)
            if s1 + s2 == 0:
                continue

            gms = strg + 'details'
            if gms in templ:
                gm = templ.get(gms).value
                gtime, rounds = get_details(gm)
                if len(rounds) > 0:
                    s1 = sum([1 for _ in rounds if _[1] > _[2]])
                    s2 = sum([1 for _ in rounds if _[1] < _[2]])
                elif s1 + s2 >= 5:
                    gtime, rounds = '', [('',s1,s2)]
                    s1 = int(s1 > s2)
                    s2 = 1 - s1
            elif s1 +s2 >= 5:
                gtime, rounds = '', [('',s1,s2)]
                s1 = int(s1 > s2)
                s2 = 1 - s1
            else:
                gtime, rounds = '', []
            #print(s1,s2,t1,t2,rounds,strb1,strb2)
            games.append((gtime,t1,t2,s1,s2,rounds))
        else:
            pass#print('uh-oh Expect once')
    return games
```
